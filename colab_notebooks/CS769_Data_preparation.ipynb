```python
all_labels_test = []
for idx in range(len(dataset['test'])):
  all_labels_test.append(dataset['test'][idx]['sentiment'])
Counter(all_labels_test)
```

```python
with open('data/twitter_sentiment140/train.tsv', "w") as f:
  for idx in range(len(dataset['train'])):
    f.write(dataset['train'][idx]['text']+"\t"+str(dataset['train'][idx]['sentiment'])+"\n")
```

```python
with open('data/twitter_sentiment140/test.tsv', "w") as f:
  for idx in range(len(dataset['test'])):
    f.write(dataset['test'][idx]['text']+"\t"+str(dataset['test'][idx]['sentiment'])+"\n")
```
